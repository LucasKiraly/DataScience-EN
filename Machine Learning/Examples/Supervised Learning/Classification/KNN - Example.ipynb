# KNN - Breast Cancer

This example has the objective to demonstrate by a simple form the use of a KNN model to predict breast cancer diagnostic

The dataset used in this project was the "Breast Cancer Wisconsin (Diagnostic) " from Scikit-Learn datasets.

# Libraries

```python
import pandas as pd

import plotly.graph_objects as go

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

# Data

### Importing data

```python
load_data = datasets.load_breast_cancer()

data = pd.DataFrame(load_data.data, columns=load_data.feature_names)

data.head()
```

### Standardizing the data

```python
scaler = StandardScaler().fit_transform(data)

standardized_data = pd.DataFrame(scaler,columns=data.columns)

standardized_data['target'] = load_data.target

standardized_data.head()
```

# Defining variables

```python
X = standardized_data.drop('target', axis=1)

y = standardized_data['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
```

# KNN model

### Searching for the best K value

```python
accuracy_rate = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    cv_score = cross_val_score(knn, X, y, cv=5)
    accuracy_rate.append(cv_score.mean())
```

```python
size = [i for i in range(1,40)]

fig_best_k = go.Figure()

fig_best_k.add_trace(go.Scatter(x=size, y=accuracy_rate, mode='lines+markers'))

fig_best_k.update_layout(xaxis_title='Number of neighbors', yaxis_title='Score', width=900, height=550)

fig_best_k.show('png')
```

### Creating model with the best K value

```python
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

### Evaluating model

```python
print(classification_report(y_test, y_pred))
```
