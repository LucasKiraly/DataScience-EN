# SVC - Breast Cancer

This example have the objective to demonstrate by a simple form the use of a SVC model to predict breast cancer diagnostic

The dataset used in this project was the "Breast Cancer Wisconsin (Diagnostic) " from Scikit-Learn datasets.

# Libraries

```python
import pandas as pd

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

# Data

### Importing data

```python
load_data = datasets.load_breast_cancer()

data = pd.DataFrame(load_data.data, columns=load_data.feature_names)

data.head()
```

### Standardizing the data

```python
scaler = StandardScaler().fit_transform(data)

standardized_data = pd.DataFrame(scaler,columns=data.columns)

standardized_data['target'] = load_data.target

standardized_data.head()
```

# Defining variables

```python
X = standardized_data.drop('target', axis=1)

y = standardized_data['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

# SVC model

### Creating model

```python
svc = SVC(C = 10, kernel = 'linear').fit(X_train, y_train)

y_pred = svc.predict(X_test)
```

### Evaluating model

```python
print(classification_report(y_test, y_pred))
```
